```python
dataset.pivot_table(values='Item_Outlet_Sales' , index='Outlet_Type')
```

From the table above it is clear that to increase  efficiency of dataset for analysis we won't be able to join these two (Supermarket Type1 and Supermarket Type2 or with Supermarket Type3) because there values noticiable  difference in between them.

```python
dataset.head(10)
```

There are '0' in Item_Visibility column which doesnot make any sense therefore we replace '0' with avg. of the column to make our dataset moore efficient.

```python
avg_Item_Visibility = dataset['Item_Visibility'].mean()
```

```python
avg_Item_Visibility
```

```python
dataset['Item_Visibility'] = dataset['Item_Visibility'].replace(0,avg_Item_Visibility)
```

```python
dataset['Item_Visibility']
```

```python
dataset['Item_Type_New'] = dataset['Item_Identifier'].apply(lambda x: x[0:2])

# now renaming the new assigned into a different cluster
dataset['Item_Type_New'] = dataset['Item_Type_New'].map({'FD':'Food',
                                                  'NC':'Non-Consumable',
                                                  'DR':'Drinks'})
                                                  
dataset['Item_Type_New'].value_counts()
```

```python
dataset.head(10)
```

We now have to calculate popularity of the mart to corresponding years

```python
dataset['Outlet_Years'] = 2013-dataset['Outlet_Establishment_Year']
```

```python
dataset['Outlet_Years'].describe()
```

```python
dataset.columns
```

```python
# convert categorical data to  numeric data
from sklearn import preprocessing
label_en= preprocessing.LabelEncoder()

#fiting label encoder using fit_transform()
#new column Outlet is taken here
dataset['Outlet'] = label_en.fit_transform(dataset['Item_Identifier'])

mod = ['Item_Fat_Content','Outlet','Outlet_Size','Item_Type_New','Outlet_Location_Type']

label_en = preprocessing.LabelEncoder()
for i in mod:
    dataset[i]= label_en.fit_transform(dataset[i])
```
