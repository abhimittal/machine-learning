```python
import numpy as np
x = np.random.rand(10)
print(x)
y = np.mean(x)
print("mean=",y)
```

```python
import numpy as np
a = np.random.rand(20,1)
b = np.var(a)
c = np.std(a)
print(a)
print("variance=",b)
print("Standard Deviation=",c)
```
